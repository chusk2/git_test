## 쯈u칠 porcentaje de pasajeros sobrevivi칩?

Para calcular la cantidad de pasajeros supervivientes, har칠 uso del m칠todo `.value_counts(normalize=True)` y a침adir칠 el par치metro `normalize` para obtener la proporci칩n de cada uno de los valores.

````python
survivors_proportions = df.Survived.value_counts(normalize=True) * 100
survivors_proportions = survivors_proportions.round(2)
survivors_proportions.index = ['Fallecidos', 'Supervivientes']
survivors_proportions.name = 'Distribuci칩n de supervivientes'
survivors_proportions
````

## 쮺u치l es la edad media de los pasajeros? 

Mediante un slicing del passenger_class_distribset, obtenemos la columna `age` y aplicamos sobre la `pandas.Series` resultante el m칠todo `.mean()`

## 쮺u치ntos pasajeros hab칤a en cada clase?

De forma similar a la 1a pregunta, hago un conteo de los diferentes valores existentes en la variable `Pclass`. Adem치s, ordeno los resultados usando los valores del 칤ndice de la serie resultante.

## 쮺u치l es la edad media de los pasajeros que sobrevivieron y de los que no?

En este caso es necesario hacer una agrupaci칩n de los datos por la variable `Survived`. A continuaci칩n, obtener la variable `Age`. Sobre los datos de esta serie agrupada, aplicar la media y un redondeo para obtener unos datos con un formato m치s adecuado.

````python
passengers_grouped_survival = df.groupby('Survived').Age
passengers_grouped_survival_ages = passengers_grouped_survival.mean().round(1)
````

Una vez obtenidos los datos, procedo a formatearlos para una apariencia m치s informativa.

````python
passengers_grouped_survival_ages.index = ['Fallecidos', 'Supervivientes']
passengers_grouped_survival_ages.name = 'Edad media en funci칩n de su supervivencia'
passengers_grouped_survival_ages
````

## 쮺u치l es la edad m치xima de los pasajeros que sobrevivieron y de los que no?

1. Reutilizo la informaci칩n de la distribuci칩n de edades de los pasajeros en funci칩n de su superviviencia de la pregunta anterior.  
2. Para obtener tanto la edad m칤nima como la m치xima, har칠 uso de las funciones de agregaci칩n `max()` y `min()`.  
3. Para aplicar dos o m치s funciones de agregaci칩n sobre una serie de datos, hago uso del m칠todo `.agg()`  
4. Formateo los valores del dataframe resultante para unos resultados m치s informativos.

````python
max_min_ages_survival_distribution = passengers_grouped_survival.agg(['min', 'max'])
````
